```python
#!pwd

#import opendpd

#opendpd.create_dataset('/content/drive/MyDrive/OpenDPD/datasets/Doherty1GHz.csv','/content/drive/MyDrive/OpenDPD/datasets/', 'Doherty1GHz', input_signal_fs=1e9, bw_main_ch=130e6, nperseg=2560)
opendpd.create_dataset('/content/drive/MyDrive/OpenDPD_firstAttempt/datasets/GaN.csv','/content/drive/MyDrive/OpenDPD_firstAttempt/datasets/', 'GaN', input_signal_fs=1e9, bw_main_ch=130e6, nperseg=2560)

print("=" * 80)
print("Attmept 1: Training a PA Model on GaN PA Data")
print("=" * 80)

pa_results = opendpd.train_pa(
    dataset_name ='/content/drive/MyDrive/OpenDPD_firstAttempt/datasets/GaN',
    PA_backbone='rvtdcnn',
    PA_hidden_size=330,
    n_epochs=50,
    lr=5e-4,
    batch_size=1024,
    accelerator='cuda',  # switch to 'cuda' or 'mps' if available
    seed=0
)

print(f"\n✓ PA model saved at: {pa_results['model_path']}")
print(f"✓ Training log saved at: {pa_results['log_path']}")
```

```python
dpd_results = opendpd.train_dpd(
    dataset_name='/content/drive/MyDrive/OpenDPD_firstAttempt/datasets/GaN',
    DPD_backbone='gmp', #DPD model definition
    DPD_hidden_size=52,  #DPD hiden size
    PA_backbone='lstm',  # PA model trained above
    PA_hidden_size=54,  # PA model trained above
    n_epochs=50,
    lr=5e-3,
    batch_size=1024,
    accelerator='cuda',
    seed=0,
    thx=0.0,  # Threshold for input deltas (for delta-based models)
    thh=0.0   # Threshold for hidden state deltas
)

print(f"\n✓ DPD model saved at: {dpd_results['model_path']}")
print(f"✓ Training log saved at: {dpd_results['log_path']}")
```

```python
run_results = opendpd.run_dpd(
    dataset_name='/content/drive/MyDrive/OpenDPD_firstAttempt/datasets/GaN',
    PA_backbone='lstm',
    PA_hidden_size=128,
    DPD_backbone='lstm',
    DPD_hidden_size=128,
    accelerator='cuda'
)

print(f"\n✓ DPD output saved at: {run_results['output_path']}")
```
